```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('Playground.csv')
df
```

```python
df.info()
```

```python
df['Date']=pd.to_datetime(df['Date'])
df.info()
```

```python
df['day']=df.Date.dt.day
df['month']=df.Date.dt.month
df['year']=df.Date.dt.year
df
```

```python
df=df.astype({"first_purchase_os_name":'category'})
df=df[df['first_purchase_os_name'].notnull()]
df.info()
```

```python
df=df.sort_values(by='Date',ascending=True)
df
```

```python
df['year_month']=df['year'].astype(str) +"_"+ df['month'].astype(str)
df=df[df['year_month']!='2022_1']
df
```

```python
yearMonthList=pd.unique(df['year_month'])
yearMonthList
```

```python
mac_df=df[df['first_purchase_os_name'].str.contains('Mac')]
mac_df=mac_df.groupby(by='year_month').sum()
mac_df
```

```python
windows_df=df[df['first_purchase_os_name'].str.contains('Windows')]
windows_df=windows_df.groupby(by='year_month').sum()
windows_df
```

```python
android_df=df[df['first_purchase_os_name'].str.contains('Android')]
android_df=android_df.groupby(by='year_month').sum()
android_df
```

```python
ios_df=df[df['first_purchase_os_name'].str.contains('iOS')]
ios_df=ios_df.groupby(by='year_month').sum()
ios_df
```

```python
ig=plt.figure(figsize=(12,8), dpi= 100, facecolor='w', edgecolor='k')
plt.plot(yearMonthList,mac_df['totalOS'], label="Mac")
plt.plot(yearMonthList,windows_df['totalOS'], label="Windows")
plt.xticks(np.arange(0,len(yearMonthList)+4,4),rotation='vertical')
plt.legend()
plt.show()
```

```python
ig=plt.figure(figsize=(12,8), dpi= 100, facecolor='w', edgecolor='k')
plt.plot(yearMonthList,android_df['totalOS'], label="Android")
plt.plot(yearMonthList,ios_df['totalOS'], label="iOS")
plt.xticks(np.arange(0,len(yearMonthList)+4,4),rotation='vertical')
plt.legend()
plt.show()
```
